```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(r'C:\Users\dvr_s\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv',skiprows =3)
```

```python
data.drop(columns=['Unnamed: 69','Country Code', 'Indicator Code'],inplace=True)
```

```python
data.columns
```

```python
population_2023 = data[['Country Name', '2023']].dropna()
```

```python
top10 = population_2023.sort_values('2023', ascending=False).head(10)
```

```python
top10.head()
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=top10,x='Country Name',y='2023',palette='viridis')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 most Populated Country of the Year 2023')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
population_2024 = data[['Country Name', '2024']].dropna()
```

```python
top15 = population_2024.sort_values('2024', ascending=False).head(15)
```

```python
top15
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=top15, x='2024', bins=10, kde=True, color='skyblue')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.title('Distribution of Population of Top 15 Countries (2024)')
plt.grid(axis='y')
plt.show()
```

# The histogram shows the distribution of population (a continuous variable) for the Top 15 most populated countries in a selected year.

# The bar chart shows the Top 10 most populated countries, with country names (categorical variable) on the x-axis and their population on the y-axis for the selected year.
